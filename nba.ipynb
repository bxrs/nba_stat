```python
from selenium import webdriver
from selenium.webdriver.common.by import By  

driver=webdriver.Chrome()
driver.get("https://www.nba.com/games")

team=driver.find_elements(By.CLASS_NAME, "MatchupCardTeamName_teamName__9YaBA")
record=driver.find_elements(By.CLASS_NAME, "MatchupCardTeamRecord_record__20YHe")
# game_leaders=driver.find_elements(By.CLASS_NAME, "GameCardLeaders_gclTable__iiNs9")
# game_score=driver.find_elements(By.CLASS_NAME, "GameCardMatchup_statusWrapper__TDbQz")
game_leaders=driver.find_elements(By.CLASS_NAME, "GameCardLeaders_gclTable__iiNs9")
```

```python
for i in range(len(team)):
    print(f"{team[i].text}: {record[i].text}")

teams=[]
wins = []
losses = []

for i in range(len(team)):
    t_name=team[i].text
    rec=record[i].text
    win, loss =map(int, rec.split('-'))
    teams.append(t_name)
    wins.append(win)
    losses.append(loss)

driver.quit()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sorted_indices = sorted(range(len(wins)), key=lambda i: wins[i], reverse=True)
sorted_teams = [teams[i] for i in sorted_indices]
sorted_wins = [wins[i] for i in sorted_indices]
sorted_losses=[losses[i] for i in sorted_indices]

plt.bar(sorted_teams, sorted_wins, color='blue', label='Team Wins')
plt.bar(sorted_teams, sorted_losses, bottom=sorted_wins, color='red', label='Team Losses')
plt.title('NBA Teams Wins')
plt.ylabel('Games')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.legend()
plt.show()
```
